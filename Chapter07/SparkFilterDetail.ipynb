```
# exercise to show transformation of map()
print("This exercise is to show map of finding cube of a number")
inDataStep1 = sc.parallelize([1,2,3,4,5])
```

```
#the function takes input number and finds the cube of the number
outDataStep1 = inDataStep1.map(lambda x: x*x*x).collect()
```

```
#printing the output data set
for x in outDataStep1:
  print (x)  
```

```
#take the outData as the input for the filter function for find odd numbers
print("This exercise is to show filter of finding odd numbers")
#the input data is parallelized
inDataStep2 = sc.parallelize(outDataStep1)
```

```
#the filter for funding the odd values
outDataStep2 = inDataStep2.filter(lambda x: x%2!=0).collect()
```

```
#the result values of filter are printed
for x in outDataStep2:
  print (x)
```

```
#take the outData of map as the input for the filter function for finding even numbers
print("This exercise is to show filter of finding even numbers")
inDataStep3 = sc.parallelize(outDataStep1)
```

```
#applying the filter to finding even value
outDataStep3 = inDataStep2.filter(lambda x: x%2==0).collect()
```

```
#looping through the dataset
for x in outDataStep3:
  print (x)
```

```
#finding the union between the even and odd filters
print("This exercise is to show union of finding odd and even filter outputs")
inDataStep4 = sc.parallelize(outDataStep2)
inDataStep5 = sc.parallelize(outDataStep3)
```

```
#doing an union between odd and even values and collecting it
outDataStep6 = sc.union([inDataStep4, inDataStep5]).collect()
```

```
#looping through the dataset
for x in outDataStep6:
  print (x)
```

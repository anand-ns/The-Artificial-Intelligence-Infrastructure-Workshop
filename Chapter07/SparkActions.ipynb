```
inString = ["Spark", "is", "a", "powerful", "cluster", "based", "parallel", "data", "processing", "framework"]
wordsString = sc.parallelize(inString)
words = wordsString.collect()
for word in words:
  print(word)
```

```
# counts the number of elements in a dataset
wordsString.count()
```

```
# returns the first element 
wordsString.first()
```

```
#takes the first n elements of the dataset
wordsString.take(3)
```

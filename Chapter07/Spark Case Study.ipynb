```
# File location and type
file_location = "/FileStore/tables/Film_Locations_in_San_Francisco.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
dataTable = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(dataTable)
```

```
# Create a view or table
temp_table_name = "Film_Locations_in_San_Francisco_New"
dataTable.createOrReplaceTempView(temp_table_name)
```
